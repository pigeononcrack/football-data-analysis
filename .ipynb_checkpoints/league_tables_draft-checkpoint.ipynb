```python
import sqlite3
import numpy as np
import pandas as pd
```

```python
conn = sqlite3.connect('database.sqlite')
```

```python
pd.read_sql('SELECT * FROM match LIMIT 1',conn).columns.tolist()
```

```python
matches_df = pd.read_sql(
    "SELECT league_id, season, match_api_id, home_team_api_id, away_team_api_id, home_team_goal, away_team_goal FROM match",
    conn, index_col = "match_api_id")
    
```

```python
matches_df.info()
```

```python
matches_df.head()
```

```python
conditions = [
    (matches_df["home_team_goal"]>matches_df["away_team_goal"]),
    (matches_df["home_team_goal"]<matches_df["away_team_goal"]),
    (matches_df["home_team_goal"]==matches_df["away_team_goal"])
]

choices = ['home_win', 'away_win', 'draw']

matches_df['result'] = np.select(conditions, choices, default='Unknown')
```

```python
home_points = epl_2009_standings.groupby("home_team_api_id")["home_team_pts"].sum()
```

```python
home_points
```

```python
home_points.info()
```

```python
away_points = epl_2009_standings.groupby("away_team_api_id")["away_team_pts"].sum()
```

```python
away_points
```

```python
total_points = away_points + home_points
```

```python
total_points.name = "points"
```

```python
total_points
```

```python
home_goals_scored = epl_2009_standings.groupby("home_team_api_id")["home_team_goal"].sum()
```

```python
away_goals_scored = epl_2009_standings.groupby("away_team_api_id")["away_team_goal"].sum()
```

```python
total_goals_scored = home_goals_scored + away_goals_scored
```

```python
total_goals_scored.name = "goals scored"
```

```python
total_goals_scored
```

```python
home_goals_conceded = epl_2009_standings.groupby("home_team_api_id")["away_team_goal"].sum()
```

```python
away_goals_conceded = epl_2009_standings.groupby("away_team_api_id")["home_team_goal"].sum()
```

```python
total_goals_conceded = home_goals_conceded + away_goals_conceded
```

```python
total_goals_conceded.name = "goals conceded"
```

```python
total_goals_conceded
```

```python
home_record = epl_2009_standings.groupby("home_team_api_id")["result"].value_counts().unstack().fillna(0).astype(int)
```

```python
away_record = epl_2009_standings.groupby("away_team_api_id")["result"].value_counts().unstack().fillna(0).astype(int)
```

```python
home_record
```

```python
away_record
```

```python
epl_table = pd.concat([total_points, total_goals_scored, total_goals_conceded], axis=1)
```

```python
epl_table
```

```python
epl_table["wins"] = home_record["home_win"] + away_record["away_win"]
```

```python
epl_table
```

```python
epl_table["draws"] = home_record["draw"] + away_record["draw"]
```

```python
epl_table["losses"] = home_record["away_win"] + away_record["home_win"]
```

```python
epl_table
```

```python
epl_table["club"] = epl_table.index.map(teams_df["team_long_name"])
```

```python
epl_table["goal_difference"] = epl_table["goals scored"] - epl_table["goals conceded"]
```
