## 1.Average goal's scored per league

The first thing we need to do is to create dataframe from <matches> table using pandas library. There are a lot of useless info for our task, so we're just gonna get columns that are needed for our task. Firstly I'm going to check all of the columns in the table, then call needed data into dataframe.

```python
pd.read_sql("SELECT * FROM Match LIMIT 1", conn).columns.tolist()
```

Here we can see that our table has loads of unneccessary columns for this task. So we're just gonna stick with a couple of them:
1. match_api_id
2. league_id
3. season
4. home_team_goal
5. away_team_goal

```python
matches_df = pd.read_sql("SELECT match_api_id, league_id, season, home_team_goal, away_team_goal FROM Match", conn)
```

```python
matches_df.head()
```

The next step is going to be the creation of new column with combined goals scored in every match. That's pretty easy to do:

```python
matches_df["total_goals"] = matches_df["home_team_goal"]+matches_df["away_team_goal"]
```

```python
matches_df.head()
```

We can see that there are a new column and based on 5 matches it is counting correctly. Hurray! The next step would be to use the groupby pandas method to group goals per league and find mean of total goals scored per league 

```python
matches_df.groupby("league_id")["total_goals"].mean()
```

Another win! But we need to make league name to appear instead of id's, to make it more user-friendly. So no we're gonna use <league> table, where needed data is stored

```python
league_df=pd.read_sql("SELECT id, name FROM League",conn, index_col = "id")
```

```python
league_df.head()
```

Now that we have another dataframe with league names we can merge our two dataframes together, using the mutual column in both tables - id and league_id and a merge method from pandas library

```python
matches_df = matches_df.merge(league_df, left_on = "league_id", right_on = 'id')
```

```python
matches_df.head()
```

We can see that the name of the league is now in our main table. League_id column is now useless so we're gonna drop it.

```python
matches_df = matches_df.drop(["league_id"], axis = 1)
```

```python
matches_df.head()
```

Now the method groupby will look a lot more friendly:

```python
matches_df.groupby("name")["total_goals"].mean()
```

```python
matches_df = matches_df.rename(columns = {"name":"league"})
```

```python
matches_df.groupby("league")["total_goals"].mean().plot(kind="barh", title = "Average Goals Per Game in Leagues")
```

As we can see in a graph - the most goals on average is scored in Netherlands Eradivisie. Switzerland Super League, Germany 1. Bundesliga and Belgium Jupiler League is not far off too. Polish and French leagues see the least amount of goals on average. Fans in those leagues are more bored on average!

## 2. How did this number change throughout the seasons

So now we know which leagues are the goalest. Let's see how average number of goals did change throughout the seasons available in our database. Firstly - let's look at how many seasons our matches table do cover

```python
len(matches_df["season"].unique())
```

We see that it's 8 seasons. Let's make a graph using method groupby again, cause we have all the data needed

```python
sns.set_style("whitegrid")
matches_df.groupby(["season","league"])["total_goals"].mean().unstack().plot(legend=True, title="Average goals by leagues per season", figsize=(15,6))
```

Now we can see the dynamic of Goals Per Season throughtout every league in our dataset. We can see that Polish league had it really bad at 2008/2009, being absolute worst at having goals scored, but in 8 years it got itself to a lot better positiong overall. We can also see that Belgian league is having it's up's and down's in terms of the goals scored on average. Also Serie A, Italian league, seems to be the most stable and nearly haven't changed in 8 years.

Now we're gonna dive into a bit of a different data to work with and gonna find out which of the teams are the best in terms of win percentage at two different stages - home and away. We're going to use <Team> table to find out the team names that are hiding under id's in our <Match> table, merge them together and count their record throughout 8 seasons.

First of all - let's look at the <Team> table. We'll also going to check if there is any data that is missing from it, so there are no "blanks" in our dataframe.

```python
pd.read_sql("SELECT * FROM team", conn)
```

```python
pd.read_sql("SELECT * FROM team", conn).info()
```

```python
pd.read_sql("SELECT * FROM team", conn).isna().sum()
```

Here we've found 11 blanks in team_fifa_api_id column. If I had to guess - those teams haven't been in FIFA games or they have been, but under different names due to licensing issues. Let's see, out of curiosity, what are those teams

```python
teams_df = pd.read_sql("SELECT * FROM team", conn)
```

```python
teams_df[teams_df["team_fifa_api_id"].isna()]
```

Here we can see the names of those teams. Let's leave this blanks for authors of this dataset to think about cause this missing data has no impact on what we're going to do. We even gonna drop this fifa id's, cause they

```python
teams_df = teams_df.drop(["team_fifa_api_id", "team_short_name", "id"],axis = 1)
```

```python
teams_df.head()
```

Now we have two dataframes to work with for this task - matches_df and teams_df. To matches_df we need to add two columns with team participating in a match id's. Let's remind ourself's what does <match> table looks like and what are those columns called.

```python
pd.read_sql("SELECT * FROM Match LIMIT 1", conn).columns.tolist()
```

Here we can see that we're interested in columns 'home_team_api_id', 'away_team_api_id'. Let's add them to our matches_df dataframe. We need to create another dataframe with 'home_team_api_id', 'away_team_api_id' and match_ids to merge two df's together

```python
teams_match_id = pd.read_sql("SELECT match_api_id, home_team_api_id, away_team_api_id FROM match", conn)
```

```python
teams_match_id.head()
```

```python
teams_match_id.info()
```

Now we're gonna merge them together with our key - "match_api_id". In our case we've could just add two columns with team_id's, but to make it in a right way, in case there were changes to data in matches_df, we have to merge them like that - just to be sure.

```python
matches_df = matches_df.merge(
    teams_match_id, 
    on = "match_api_id", 
    how = "left"
)
```

```python
matches_df.head()
```

```python
matches_df.info()
```

We've added teams id's successfully. Now we're going to add actual names of the teams.

```python
matches_df = matches_df.merge(
    teams_df,
    left_on = "home_team_api_id",
    right_on = "team_api_id",
    how = "left"
)
```

```python
matches_df.head()
```

```python
matches_df.info()
```

```python
matches_df = matches_df.merge(
    teams_df,
    left_on = "away_team_api_id",
    right_on = "team_api_id",
    how = "left"
)
```

```python
matches_df.head()
```

```python
matches_df.info()
```

```python
matches_df = matches_df.drop(["home_team_api_id","away_team_api_id","team_api_id_x", "team_api_id_y"], axis = 1)
```

```python
matches_df.info()
```

```python
matches_df = matches_df.rename(
    columns = {"team_long_name_x":"home_team",
               "team_long_name_y":"away_team"
              })
```

```python
matches_df.head()
```

Now we only have the data we need and also some data from other tasks, but we're not bothered by it. Let's now count wins, draws and loses for each team in our table. We'll create another column in matches_df with the results of each game. Then we'll create two dataframes - home_record and away_record where we'll gather data about each team's record of their results.

```python
conditions = [
    (matches_df["home_team_goal"]>matches_df["away_team_goal"]),
    (matches_df["home_team_goal"]<matches_df["away_team_goal"]),
    (matches_df["home_team_goal"]==matches_df["away_team_goal"])
]

choices = ['home_win', 'away_win', 'draw']

matches_df['result'] = np.select(conditions, choices, default='Unknown')
```

```python
matches_df.info()
```

Now we've found the results of all 25979 matches.

```python
home_record = matches_df.groupby("home_team")["result"].value_counts().unstack().fillna(0).astype(int)
```

```python
home_record.head()
```

```python
home_record.info()
```

```python
away_record = matches_df.groupby("away_team")["result"].value_counts().unstack().fillna(0).astype(int)
```

```python
away_record.head()
```

```python
away_record.info()
```

Here we go - we've created two dataframes and we have all the data needed to find out the most succesfull teams at home turf and away

```python
home_record["win percentage, %"] = 100 * home_record["home_win"]/(home_record["home_win"] + home_record["away_win"] + home_record["draw"])
```

```python
home_record.describe()
```

```python
home_record.head()
```

```python
home_record.info()
```

Here we can see what are the most succesfull teams at the home turf. There are football giants that is Barca and Real Madrid, dominant teams in Portugese league - Porto and Benfica. And others, who traditionally play well.

```python
home_record["win percentage, %"].nlargest(15)
```

Let's do the same and check what teams play the best football not being at their home stadium.

```python
away_record["win percentage, %"] = 100 * away_record["away_win"]/(away_record["home_win"] + away_record["away_win"] + away_record["draw"])
```

```python
away_record.head()
```

```python
away_record.info()
```

```python
away_record["win percentage, %"].nlargest(15)
```

We can see lot's of familiar names - almost every club that is successful at home is more or less successful away. We can also compare the most winning teams at home and their away record. We'll create another dataframe with home and away percentage wins for 15 teams, that are in the top percentile.

```python
top15 = home_record["win percentage, %"].nlargest(15)
```

```python
top15
```

```python
top15.info()
```

```python
top15 = top15.to_frame()
```

```python
top15 = top15.reset_index()
```

```python
top15 = top15.rename(columns = 
             {"win percentage, %":"home_wins,%",
              "home_team":"team"
             })
```

```python
top15 = top15.merge(
    away_record, 
    left_on = "team",
    right_on = "away_team")
```

```python
top15.head()
```

```python
top15 = top15.drop(["away_win","draw","home_win"], axis = 1)
```

```python
top15 = top15.rename(
    columns = {
        "win percentage, %":"away_wins,%"
    })
```

```python
top15.head()
```

```python
top15.info()
```

```python
top15 = top15.set_index("team")
```

```python
top15.plot(kind="bar", title = "Home vs Away WINS", figsize = (10,8))
```

On the final graph we can see the differences between home and away record of the teams. We can see just from the graph that there are clear outliers - Rangers and Manchester City. Rangers wins at home almost just as much as away - seems their oppossition in the league are weak. And with Manchester City - they have distinct difference between playing at home and away - seems that they've really build their fortress at Etihaad.
