```python
team_ids=pd.read_sql("SELECT team_api_id, team_long_name FROM Team",conn, index_col = "team_api_id")
```

```python
match2 = pd.read_sql("SELECT id, season, match_api_id, home_team_api_id,away_team_api_id,home_team_goal,away_team_goal FROM match", conn, index_col="id")
```

```python
match2.head()
```

```python
match2 = match2.merge(team_ids, left_on = "home_team_api_id", right_on = 'team_api_id')
```

```python
match2.head()
```

```python
match2 = match2.rename(columns={"team_long_name": "home_team_name"})
```

```python
match2 = match2.merge(team_ids, left_on = "away_team_api_id", right_on = 'team_api_id')
```

```python
match2 = match2.rename(columns={"team_long_name": "away_team_name"})
```

```python
match2.head()
```

```python
match2.drop(["home_team_api_id","away_team_api_id"], axis = 1)
```

```python
match2.head()
```

```python
conditions = [
    (match2["home_team_goal"]>match2["away_team_goal"]),
    (match2["home_team_goal"]<match2["away_team_goal"]),
    (match2["home_team_goal"]==match2["away_team_goal"])
]

# 2. Define the matching choices for each condition
choices = ['home_win', 'away_win', 'draw']

# 3. Apply np.select with a fallback default value
match2['result'] = np.select(conditions, choices, default='Unknown')
```

```python
match2.head()
```

```python
match2.groupby("home_team_name")["result"]
```

```python
match2['result'].value_counts()
```

```python
match2.groupby("home_team_name")["result"].value_counts().unstack().fillna(0).astype(int)
```

```python
home_results = match2.groupby("home_team_name")["result"].value_counts().unstack().fillna(0).astype(int)
```
